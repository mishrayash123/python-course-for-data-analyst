```python
import matplotlib.pyplot as plt
```

```python
# importing matplotlib module
from matplotlib import pyplot as plt

# x-axis values
x = [5,2,9,4,7]

# y-axis values
y = [10,5,8,4,2]

# function to plot
plt.plot(x,y)

plt.show()
```

```python
# importing matplotlib module
from matplotlib import pyplot as plt

# x-axis values
x = [5,2,9,4,7]

# y-axis values
y = [10,5,8,4,2]

# function to plot
plt.plot(x,y)
plt.plot(x,y, marker = 'o')

plt.show()
```

```python
# importing matplotlib module
from matplotlib import pyplot as plt

# x-axis values
x = [5,2,9,4,7]

# y-axis values
y = [10,5,8,4,2]

# function to plot
plt.plot(x,y,marker='>')
plt.plot(y, marker = 'o')

plt.show()
```

# Bar graph

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']

plt.bar(x,y,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [5,6,7,8]
plt.title("scatter")
plt.xlabel("yash")
plt.ylabel("aditi")
plt.scatter(x,y,color="r")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [5,6,7,8]
plt.title("scatter")
plt.xlabel("yash")
plt.ylabel("aditi")
c=['red','orange','blue','green']
plt.scatter(x,y,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [5,6,7,8]
plt.title("scatter")
plt.xlabel("yash")
plt.ylabel("aditi")
plt.scatter(x,y,color="r",s=150)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["english","math","hindi","sanskrit"]

plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["english","math","hindi","sanskrit"]
c=['red','orange','blue','green']
plt.pie(x,labels=y,colors=c)
plt.legend()
plt.show()
```
